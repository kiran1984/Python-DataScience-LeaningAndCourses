```python
import pandas as pd
```

```python
melbourne_data=pd.read_csv("melb_data.csv")
melbourne_data.describe()
```

```python
melbourne_data.head()
```

```python
y=melbourne_data.Price
y
```

```python
melbourne_data=melbourne_data.drop("Price",axis=1)
melbourne_data
```

```python
cols_with_missing=[col for col in melbourne_data.columns if melbourne_data[col].isnull().any()]
cols_with_missing
melbourne_data.drop(cols_with_missing,axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(melbourne_data,y,random_state=1)
X_train
```

```python
s=[col for col in X_train.columns if X_train[col].dtypes=="object"]
print("Categorical columns:")
print(s)
```

# First Approach, droping all categorical columns

```python
drop_X_train=X_train.drop(s,axis=1)
drop_X_valid=X_valid.drop(s,axis=1)

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

model=RandomForestRegressor(random_state=1)
model.fit(drop_X_train,y_train)

val_predicted=model.predict(drop_X_valid)
print("Calculating MAE for the first approach:")
print(mean_absolute_error(val_predicted,y_valid))
```

# Second Approach, label enconding

```python
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
label_X_train=X_train.copy()
label_X_valid=X_valid.copy()
my_encoder=LabelEncoder()
s=[col for col in label_X_train.columns if label_X_train[col].dtype=="object"]
for col in s:
    label_X_train[col] = my_encoder.fit_transform(X_train[col])
    label_X_valid[col] = my_encoder.fit_transform(X_valid[col])
model=RandomForestRegressor(random_state=1)
model.fit(label_X_train,y_train)
val_predicted=model.predict(label_X_valid)
print("Calculating MAE for second approach:")
print(mean_absolute_error(val_predicted,y_valid))
```

# Third approach, onehotencoding

```python
from sklearn.preprocessing import OneHotEncoder
OH_encoder=OneHotEncoder(handle_unknown='ignore',sparse=False)
OH_cols_train=pd.DataFrame(OH_encoder.fit_transform(X_train[s]))
OH_cols_valid=pd.DataFrame(OH_encoder.transform(X_valid[s]))
OH_cols_train.index=X_train.index
OH_cols_valid.index=X_valid.index
num_X_train=X_train.drop(s,axis=1)
num_X_valid=X_valid.drop(s,axis=1)
OH_X_train=pd.concat([num_X_train,OH_cols_train],axis=1)
OH_X_valid=pd.concat([num_X_valid,OH_cols_valid],axis=1)
model=RandomForestRegressor(random_state=1)
model.fit(OH_X_train,y_train)
val_predicted=model.predict(OH_X_valid)
print("Calculating MAE for the third approach:")
print(mean_absolute_error(val_predicted,y_valid))
```
