```python
import pandas as pd
```

```python
melbourne_data=pd.read_csv("melb_data.csv")
melbourne_data.describe()
```

```python
melbourne_data.head()
```

```python
melbourne_data=melbourne_data.dropna(axis=0)
melbourne_data
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
y=melbourne_data.Price
y
```

```python
melbourne_features=['Rooms', 'Bathroom', 'Landsize', 'BuildingArea','YearBuilt', 'Lattitude', 'Longtitude']
X=melbourne_data[melbourne_features]
```

```python
train_X, val_X,train_y,val_y=train_test_split(X,y,random_state=0)
def get_mee(max_leaf_nodes,train_X,val_X,train_y,val_y):
    model=DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=0)
    model.fit(train_X,train_y)
    val_predicted=model.predict(val_X)
    MAE=mean_absolute_error(val_y,val_predicted)
    return MAE    
```

```python
for max_leaf_nodes in [5,50,500,5000]:
    mae=get_mee(max_leaf_nodes,train_X,val_X,train_y,val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes,mae))
```
