```python
import pandas as pd
```

```python
melbourne_data=pd.read_csv("melb_data.csv")
melbourne_data.describe()
```

```python
melbourne_data.head()
```

```python
from sklearn.model_selection import train_test_split
y=melbourne_data.Price
y
```

```python
melbourne_features=['Rooms', 'Bathroom', 'Landsize', 'BuildingArea','YearBuilt', 'Lattitude', 'Longtitude']
X=melbourne_data[melbourne_features]
X
```

# First Approach, dropping all missing columns

```python
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=1)
```

```python
cols_with_missing=[col for col in train_X.columns if train_X[col].isnull().any()]
cols_with_missing
```

```python
reduced_train_X=train_X.drop(cols_with_missing,axis=1)
reduced_val_X=val_X.drop(cols_with_missing,axis=1)
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
melbourne_model=RandomForestRegressor(random_state=1)
melbourne_model.fit(reduced_train_X,train_y)
val_predicted=melbourne_model.predict(reduced_val_X)
print("MAE from approch 1(Dropping all missing columns):")
print(mean_absolute_error(val_predicted,val_y))
```

# Second Approach, imputation

```python
from sklearn.impute import SimpleImputer
```

```python
my_imputer=SimpleImputer()
imputed_train_X=pd.DataFrame(my_imputer.fit_transform(train_X))
imputed_val_X=pd.DataFrame(my_imputer.fit_transform(val_X))
```

```python
imputed_train_X.columns=train_X.columns
imputed_val_X.columns=val_X.columns
```

```python
melbourne_model=RandomForestRegressor(random_state=1)
melbourne_model.fit(imputed_train_X,train_y)
val_predicted=melbourne_model.predict(imputed_val_X)
print("MAE from approch 2(Impute the missing values):")
print(mean_absolute_error(val_predicted,val_y))
```

# Third Approach, extension to imputation

```python
train_X_plus=train_X.copy()
val_X_plus=val_X.copy()
```

```python
for col in cols_with_missing:
    train_X_plus[col+" was missing"]=train_X_plus[col].isnull()
    val_X_plus[col+" was missing"]=val_X_plus[col].isnull()
```

```python
imputed_train_X_plus=pd.DataFrame(my_imputer.fit_transform(train_X_plus))
imputed_val_X_plus=pd.DataFrame(my_imputer.fit_transform(val_X_plus))
imputed_train_X_plus.columns=train_X_plus.columns
imputed_val_X_plus.columns=val_X_plus.columns
```

```python
melbourne_model=RandomForestRegressor(random_state=1)
melbourne_model.fit(imputed_train_X_plus,train_y)
```

```python
val_predicted=melbourne_model.predict(imputed_val_X_plus)
print("MAE from approach 3(extension to imputation):")
print(mean_absolute_error(val_predicted,val_y))
```
